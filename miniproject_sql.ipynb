# SQL mini project
For this SQL miniproject, you will work with the "country_club" database. 
The SQL file to create the database has been downloaded and stored under current directory with the file name as 'country_club.sql'. This database contains 3 tables:
    i) the "Bookings" table,
    ii) the "Facilities" table, and
    iii) the "Members" table.

In this mini project, there are a series of questions. 
Please see the cell right after each question for the solution submitted. 

```python
#import sqlite3 module, create an empty sqlite3 database
import sqlite3
conn=sqlite3.connect('country_club.db')
c=conn.cursor()

#readin SQL commands from SQL file named 'country_club.sql' into a single string, skipping lines starting with '--', 'SET' and '/*'
f = open('country_club.sql')
sqlfile=''
for line in f:
    if line.startswith('--') or line.startswith('SET') or line.startswith('/*'):
        continue
    sqlfile += line
f.close()

#execute commands from the input file on at a time and fill in the 'country_club.db' database
commands = sqlfile.split(';')
for command in commands:
    try: 
        c.execute(command)
    except:
        print ("Command skipped:",command)
```
